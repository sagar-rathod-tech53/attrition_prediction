```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../datasets/hr_data.csv")
```

```python
print(data.shape)
```

```python
categorical_cols = [
    'Gender',
    'OverTime',
    'Attrition'
]

for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

```python
data.drop(
    columns=['Over18', 'EmployeeCount', 'StandardHours', 'EmployeeNumber'],
    inplace=True,
    errors='ignore'
)

X = data.drop('Attrition', axis=1)
y = data['Attrition']

X = pd.get_dummies(X, drop_first=True)

X = X.astype(float)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
X_columns = X.columns   # original before scaling/splitting
```

```python
X_train_df = pd.DataFrame(X_train, columns=X_columns)
X_test_df = pd.DataFrame(X_test, columns=X_columns)
```

```python
y_train_df = pd.DataFrame(y_train, columns=["Attrition"])
y_test_df = pd.DataFrame(y_test, columns=["Attrition"])
```

```python
print(type(X_train))
print(type(X_test))
print(type(y_train))
print(type(y_test))
```

```python
X_train_df.to_csv("../datasets/X_train.csv", index=False)
X_test_df.to_csv("../datasets/X_test.csv", index=False)

y_train_df.to_csv("../datasets/y_train.csv", index=False)
y_test_df.to_csv("../datasets/y_test.csv", index=False)
```

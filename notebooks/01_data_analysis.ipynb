```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../datasets/hr_data.csv")
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
print(data.duplicated().sum())
```

```python
print(data['Attrition'].value_counts())
print(data['Attrition'].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Attrition',
    data=data,
    palette='Set2'
)

plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,6))

data['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['skyblue','salmon']
)

plt.ylabel("")
plt.title("Attrition Percentage Distribution")

plt.show()
```

```python
sns.histplot(data['Age'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data['MonthlyIncome'],
    bins=30,
    kde=True,
    color='green'
)

plt.title("Monthly Income Distribution")

plt.show()
```

```python
sns.histplot(
    data['DistanceFromHome'],
    bins=20,
    color='orange'
)

plt.title("Distance From Home")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='Age',
    data=data
)

plt.title("Age vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=data
)

plt.title("Monthly Income vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=data
)

plt.title("OverTime vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    x='JobRole',
    hue='Attrition',
    data=data
)

plt.xticks(rotation=45)

plt.title("Job Role vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='MaritalStatus',
    hue='Attrition',
    data=data
)

plt.title("Marital Status vs Attrition")

plt.show()
```

```python
numeric_df = data.select_dtypes(
    include=['int64','float64']
)

plt.figure(figsize=(15,10))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.title("Correlation Matrix")

plt.show()
```

```python
data['Attrition'] = data['Attrition'].map({
    'Yes':1,
    'No':0
})

corr = data.corr(numeric_only=True)

corr['Attrition'].sort_values(
    ascending=False
)

plt.figure(figsize=(8,6))

corr['Attrition'].sort_values().plot(
    kind='barh'
)

plt.title("Feature Correlation with Attrition")

plt.show()
```
